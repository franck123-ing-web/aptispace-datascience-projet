détecter les valeurs manquantes
obtenir des premiers insights sur les prix des maisons

1. Chargement du dataset

Dans un premier temps, on télécharge et extrait le dataset depuis Kaggle

Objectif :
récupérer les données officielles

garantir une structure reproductible du projet

2. Chargement dans Pandas

Objectif :
charger les données dans un DataFrame Pandas
préparer l’analyse exploratoire

3. Analyse générale du dataset

Ce que ça permet :

4. Analyse des valeurs manquantes

Objectif :
identifier les colonnes avec données manquantes
comprendre la qualité du dataset

5. Insights rapides sur la variable cible

 Ce que ça apporte :
comprendre la distribution globale des prix
détecter si les prix sont dispersés ou concentrés
poser une base pour la modélisation

```python
import pandas as pd
import os
import zipfile

# ==============================
# DOWNLOAD DATASET
# ==============================

os.system(
    "kaggle competitions download -c house-prices-advanced-regression-techniques"
)

# ==============================
# EXTRACTION
# ==============================

with zipfile.ZipFile(
    "house-prices-advanced-regression-techniques.zip",
    "r"
) as zip_ref:
    zip_ref.extractall("data/raw")

# ==============================
# LOAD DATASET
# ==============================

df = pd.read_csv("data/raw/train.csv")

# ==============================
# EDA - BASE INFO
# ==============================

print("📊 Dimensions :", df.shape)

print("\n📌 Infos générales :")
print(df.info())

print("\n📈 Statistiques descriptives :")
print(df.describe())

# ==============================
# MISSING VALUES ANALYSIS
# ==============================

missing_values = df.isnull().sum()

missing_values = missing_values[missing_values > 0]

missing_values = missing_values.sort_values(ascending=False)

print("\n❗ Valeurs manquantes (top) :")
print(missing_values)

# ==============================
# BONUS INSIGHT 
# ==============================

print("\n🏠 Prix moyen des maisons :", df["SalePrice"].mean())
print("🏠 Prix médian :", df["SalePrice"].median())
print("🏠 Prix max :", df["SalePrice"].max())
print("🏠 Prix min :", df["SalePrice"].min())
```
