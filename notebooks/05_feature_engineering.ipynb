Ce notebook a pour objectif d’améliorer la qualité des données en créant de nouvelles variables (features) à partir des données existantes.

L’objectif du feature engineering est d’aider les modèles de Machine Learning à mieux comprendre les relations dans les données et ainsi améliorer la précision des prédictions.

 1. Chargement des données

Le dataset est rechargé via le pipeline officiel Kaggle afin de garantir une version cohérente et reproductible des données utilisées dans tout le projet.

 2. Création de nouvelles variables

Plusieurs nouvelles features sont créées à partir des variables existantes :

 TotalSF (surface totale)

Combine :

surface habitable (GrLivArea)
surface du sous-sol (TotalBsmtSF)

 Cette variable représente la taille globale réelle de la maison.

 HouseAge (âge de la maison)

Calculée comme :

année de vente (YrSold)
moins année de construction (YearBuilt)

 Permet de mesurer l’impact de l’ancienneté sur le prix.

 QualityScore (score de qualité global)

Produit de :

qualité générale (OverallQual)
condition générale (OverallCond)

 Donne une estimation plus globale de la qualité du bien.

 GarageScore (score garage)

Produit de :

nombre de voitures (GarageCars)
surface du garage (GarageArea)

 Permet de mieux représenter la valeur du garage dans le prix final.

 3. Aperçu des nouvelles features

Un aperçu des nouvelles variables est affiché afin de vérifier leur création et leur cohérence.

Ces features enrichissent le dataset et permettent d’apporter plus d’information aux modèles de Machine Learning.

 4. Vérification finale du dataset

Deux vérifications importantes sont réalisées :

la forme du dataset (nombre de lignes et colonnes)
les valeurs manquantes restantes

Cela permet de s’assurer que les transformations n’ont pas introduit d’erreurs dans les données

```python
import pandas as pd
import os
import zipfile

# ==============================
# LOAD DATASET (PIPELINE OFFICIEL)
# ==============================

os.system(
    "kaggle competitions download -c house-prices-advanced-regression-techniques"
)

with zipfile.ZipFile(
    "house-prices-advanced-regression-techniques.zip",
    "r"
) as zip_ref:
    zip_ref.extractall("data/raw")

df = pd.read_csv("data/raw/train.csv")

# ==============================
# FEATURE ENGINEERING
# ==============================

# Total surface
df["TotalSF"] = df["GrLivArea"] + df["TotalBsmtSF"]

# Age maison
df["HouseAge"] = df["YrSold"] - df["YearBuilt"]

# Score qualité globale
df["QualityScore"] = df["OverallQual"] * df["OverallCond"]

# Score garage
df["GarageScore"] = df["GarageCars"] * df["GarageArea"]

# ==============================
# NOUVELLES FEATURES
# ==============================

new_features = [
    "TotalSF",
    "HouseAge",
    "QualityScore",
    "GarageScore"
]

print("📊 Aperçu des nouvelles features :")
print(df[new_features].head())

# ==============================
# CHECK FINAL
# ==============================

print("\n📌 Shape dataset :", df.shape)
print("\n❗ Missing values (top 10) :")
print(df.isnull().sum().sort_values(ascending=False).head(10))
```
