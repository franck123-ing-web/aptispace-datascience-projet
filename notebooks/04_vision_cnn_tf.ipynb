### 2. Génération du jeu d'images synthétiques
Pour travailler de manière autonome sans importer de lourdes bases d'images externes, nous mettons à votre disposition cette fonction utilitaire générant des images simulées en $64 	imes 64$ pixels de route vide (Classe 0) vs route congestionnée (Classe 1).

```python
def generate_dummy_images(num_samples=150):
    images = np.zeros((num_samples, 64, 64, 3), dtype=np.float32)
    labels = np.zeros(num_samples, dtype=np.int32)
    
    for i in range(num_samples):
        label = np.random.choice([0, 1])
        labels[i] = label
        images[i, :, :, :] = 0.25 + np.random.normal(0, 0.02, (64, 64, 3))
        
        if label == 1:
            num_cars = np.random.randint(8, 20)
            for _ in range(num_cars):
                car_x = np.random.randint(5, 55)
                car_y = np.random.randint(5, 55)
                car_w = np.random.randint(4, 10)
                car_h = np.random.randint(4, 10)
                car_color = np.random.rand(3)
                images[i, car_y:car_y+car_h, car_x:car_x+car_w, :] = car_color
        else:
            num_cars = np.random.randint(0, 3)
            for _ in range(num_cars):
                car_x = np.random.randint(5, 55)
                car_y = np.random.randint(5, 55)
                car_w = np.random.randint(4, 8)
                car_h = np.random.randint(4, 8)
                car_color = np.random.rand(3)
                images[i, car_y:car_y+car_h, car_x:car_x+car_w, :] = car_color
                
    images = np.clip(images, 0.0, 1.0)
    return images, labels

X_images, y_labels = generate_dummy_images(150)
print(f"Dataset d'images généré. Dimensions : {X_images.shape}")
```

```python
# Visualisation de 4 exemples d'images synthétiques générées :
fig, axes = plt.subplots(1, 4, figsize=(10, 3))
class_names = ['Route Vide', 'Congestionné']
for i, idx in enumerate([10, 45, 90, 130]):
    axes[i].imshow(X_images[idx])
    axes[i].set_title(class_names[y_labels[idx]])
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```
