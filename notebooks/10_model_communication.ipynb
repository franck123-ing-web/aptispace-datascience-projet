Import

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("data/raw/train.csv")
```

```python
df["TotalSF"] = df["GrLivArea"] + df["TotalBsmtSF"]
df["HouseAge"] = df["YrSold"] - df["YearBuilt"]
df["QualityScore"] = df["OverallQual"] * df["OverallCond"]

features = [
    "GrLivArea",
    "OverallQual",
    "GarageCars",
    "TotalBsmtSF",
    "YearBuilt",
    "TotalSF",
    "HouseAge",
    "QualityScore"
]

df_model = df[features + ["SalePrice"]].dropna()

X = df_model[features]
y = df_model["SalePrice"]
```

##  Comprendre les données immobilières

Dans ce projet, nous cherchons à comprendre comment les caractéristiques d’une maison influencent son prix.

Les variables utilisées représentent :

- la taille de la maison
- la qualité globale
- l’année de construction
- les espaces disponibles (garage, sous-sol)

 Ces variables sont proches des critères utilisés par les agents immobiliers.

entraînement

```python
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X, y)
```

Exemple client

```python
example_house = pd.DataFrame([{
    "GrLivArea": 1500,
    "OverallQual": 7,
    "GarageCars": 2,
    "TotalBsmtSF": 800,
    "YearBuilt": 2005,
    "TotalSF": 2300,
    "HouseAge": 2026 - 2005,
    "QualityScore": 7 * 7
}])

prediction = model.predict(example_house)[0]

print(" Prix estimé :", f"{int(prediction):,} €")
```

##  Exemple concret

Imaginons une maison avec :

- 1500 m² de surface
- une qualité de 7/10
- 2 garages
- construite en 2005

 Le modèle estime automatiquement son prix en fonction de maisons similaires dans le dataset.

---

##  Comment le modèle pense

Le modèle ne “devine” pas.

Il compare cette maison à des milliers d’autres maisons déjà vendues et apprend des tendances du marché immobilier.

Importance des variables

```python
importance = pd.DataFrame({
    "Feature": features,
    "Importance": model.feature_importances_
}).sort_values("Importance")

print(importance)
```

##  Facteurs les plus importants

Le modèle montre que :

- la surface habitable est le facteur principal
- la qualité générale influence fortement le prix
- l’année de construction joue aussi un rôle important

 Cela correspond aux logiques du marché immobilier réel.
