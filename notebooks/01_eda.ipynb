```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ======================
# 1. LOAD DATA
# ======================
df = pd.read_csv("data/raw/train.csv")

print(df.head())
print(df.info())

# ======================
# 2. TARGET ANALYSIS
# ======================
print("Missing values:", df["SalePrice"].isnull().sum())

# distribution prix
plt.figure(figsize=(8,5))
plt.hist(df["SalePrice"], bins=50)
plt.title("Distribution des prix de vente")
plt.xlabel("Prix")
plt.ylabel("Fréquence")
plt.show()

# ======================
# 3. CORRELATION (IMPORTANT ML)
# ======================
numeric_df = df.select_dtypes(include=["int64", "float64"])

corr = numeric_df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()

# ======================
# 4. FEATURES IMPORTANTES
# ======================
top_features = corr["SalePrice"].sort_values(ascending=False).head(10)
print("Top features corrélées avec le prix:")
print(top_features)

# ======================
# 5. SIMPLE RELATION VISUELLE
# ======================
plt.figure(figsize=(8,5))
plt.scatter(df["GrLivArea"], df["SalePrice"])
plt.title("Surface vs Prix")
plt.xlabel("Surface habitable")
plt.ylabel("Prix")
plt.show()

# ======================
# 6. BASIC CLEANING
# ======================
df = df.dropna(subset=["SalePrice"])

print("Dataset final shape:", df.shape)
```
