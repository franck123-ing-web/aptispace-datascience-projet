Dans ce notebook, nous allons :
- télécharger le dataset officiel Kaggle (House Prices)
- extraire les fichiers
- charger les données dans un DataFrame pandas
- effectuer une première exploration rapide du dataset

On utilise la Kaggle CLI pour télécharger automatiquement le dataset officiel de la compétition **House Prices**.

Cela permet de garantir que les données sont :
- toujours à jour
- identiques à celles utilisées en compétition
- reproductibles

Le dataset est téléchargé sous forme d’archive ZIP.

Nous l’extrayons dans le dossier `data/raw/` afin de structurer proprement le projet et séparer :
- données brutes (raw)
- notebooks
- modèles

Nous chargeons le fichier `train.csv` dans un DataFrame pandas.

Ce fichier contient les caractéristiques des maisons ainsi que leur prix de vente (`SalePrice`).

 Première exploration du dataset

Cette étape permet de comprendre la structure des données :

- nombre de lignes et colonnes
- liste des variables disponibles
- aperçu des premières observations
- quantité de valeurs manquantes

Cela constitue la base de toute analyse exploratoire (EDA).

```python
import pandas as pd
import os
import zipfile

# ==============================
# DOWNLOAD DATASET
# ==============================

os.system(
    "kaggle competitions download -c house-prices-advanced-regression-techniques"
)

# ==============================
# EXTRACTION
# ==============================

with zipfile.ZipFile(
    "house-prices-advanced-regression-techniques.zip",
    "r"
) as zip_ref:
    zip_ref.extractall("data/raw")

# ==============================
# LOAD DATASET
# ==============================

df = pd.read_csv("data/raw/train.csv")

# ==============================
# DATA OVERVIEW
# ==============================

print("📊 Dataset chargé avec succès")
print("Shape :", df.shape)

print("\n📁 Colonnes disponibles :")
print(df.columns)

print("\n🔎 Aperçu des données :")
print(df.head())

print("\n❗ Valeurs manquantes globales :", df.isnull().sum().sum())
```
